```python
import numpy as np
#1.arthiemtic operations
arr1=np.array([1,2,3,4,5])
arr2=np.array([7,8,9,10,6])
print(arr1+arr2)  #element-wise operations
print(arr1-arr2)
print(arr1*arr2)
print(arr1/arr2)
print(arr1**arr2)
print(arr1%2)

#broadcasting
print(arr1+10)  #arr1+[10,10,10,10,10]-automatically does it
print(arr2*2)
```

```python
#2.matrix opearations
arr_mat1=np.array([[1,2],[4,5]])
arr_mat2=np.array([[7,8],[10,11]])
print(arr_mat1*arr_mat2)             #element-wise multiply
print(np.matmul(arr_mat1,arr_mat2))   #matrix mutiplication  (arr_mat1@arr_mat2 is the modern way)
print(np.dot(arr_mat1,arr_mat2))      #old way

print(arr_mat1.T)                #transpose 
print(np.linalg.det(arr_mat2))   #determinant
print(np.linalg.inv(arr_mat1))   #inverse
print(np.linalg.eig(arr_mat2))   #eigen values and eigen vectors
```

```python
#3.statistical opeartions
arr3=np.array([[1,5,56],[2,45,96],[85,75,92]])
print(np.sum(arr3))        #sum of all elements
print(np.sum(arr3,axis=0))  #sum of cols(axis=0)
print(np.sum(arr3,axis=1))  #sum of rows(axis=1)
print(np.mean(arr3))        #mean
print(np.std(arr3))         #standard deviation
print(np.var(arr3))         #variance
print(np.min(arr3))         #minimum value
print(np.max(arr3))         #maximum value
print(np.median(arr3))      #median value

#cumulative operations
print(np.cumsum(arr3))       #cumulative sum
print(np.cumprod(arr3))      #cumulative product
```
